# <font color='orange'>reshaping dataframe</font>

```python
# DT = pd.read_csv('../../datasets/soccer.csv', index_col='date', parse_dates=True)
DT = pd.read_csv('../../datasets/soccer.csv')

DT = DT.sort_index()
# DT = DT[DT.index.duplicated()]
DT = DT.drop_duplicates(subset='date')
DT.head()
```

```python
london_sub = DT[DT['city']=='London']
london_sub = london_sub.groupby(['tournament']).head(-1)
london_sub
```

```python
# reshape methods: pivot, pivot_table, melt
```

```python
london_sub.pivot(columns='tournament',values='home_score', index='date').plot(figsize=(15,8))
```

```python
london_sub.pivot_table(
    values='home_score',index='date',columns='tournament',margins=True,
    aggfunc='sum'
)
```

```python
pivoted = DT.pivot(columns='tournament',values='home_score', index='date').reset_index()
pivoted.melt(id_vars='date', value_name='my values').dropna().sort_values('date')
```

# <font color='orange'>merge dataframe</font>

```python
# join, merge, concat
from IPython.display import display

DT1 = pd.DataFrame({
    'key':['k0','k1','k2','k3','k4','k5'],
    'A':['A0','A1','A2','A3','A4','A5']
})

DT2 = pd.DataFrame({
    'key':['k0','k1','k2'],
    'B':['B0','B1','B2']
})

display(DT1.head())
display(DT2.head())
```

```python
DT1.set_index('key').join(DT2.set_index('key'), lsuffix='_left', rsuffix='_right', on='key')
```

```python
DT1.set_index('key').join(DT2.set_index('key'), lsuffix='_left', rsuffix='_right', on='key', how='right')
```

```python
DT1.set_index('key').join(DT2.set_index('key'), lsuffix='_left', rsuffix='_right', on='key', how='inner')
```

```python
DT1.set_index('key').join(DT2.set_index('key'), lsuffix='_left', rsuffix='_right', on='key', how='outer')
```

```python
DT1 = pd.DataFrame({
    'id':['k0','k1','k2','k3','k4','k5'],
    'name':['A0','A1','A2','A3','A4','A5'],
    'phone':['A0','A1','A2','A3','A4','A5'],
})

DT2 = pd.DataFrame({
    'user_id':['k0','k1','k2'],
    'balance':[1000,2000,3000]
})

```

```python
DT1.merge(DT2, left_on='id', right_on='user_id')
```

```python
DT1.merge(DT2, left_on='id', right_on='user_id', how='left')
```

```python
DT1.merge(DT2, left_on='id', right_on='user_id', how='inner')
```

```python
DT1.merge(DT2, left_on='id', right_on='user_id', how='outer')
```

```python
DT1.merge(DT2, left_on='id', right_on='user_id', how='right')
```

```python
DT1 = pd.DataFrame({
    'id':['A','B','C','B','C','B'],
    'name':['A0','A1','A2','A3','A4','A5'],
    'phone':['A0','A1','A2','A3','A4','A5'],
})

DT2 = pd.DataFrame({
    'user_id':['A','B','C'],
    'balance':[1000,2000,3000]
})
```

```python
pd.concat([DT1,DT2])
```

```python
pd.concat([DT1,DT2], keys=['id','user_id'])
```
