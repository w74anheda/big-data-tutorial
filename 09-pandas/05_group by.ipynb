# group by
 * aggregation
 * transform
 * filteration

```python
import pandas as pd
import numpy as np
```

```python
DT = pd.read_csv('../../datasets/titanic.csv')
DT.head()
```

```python
DT.to_string(index=False)
```

```python
DT.groupby('Pclass').groups
DT.groupby('Pclass').groups[1] # group item
DT.groupby('Pclass').groups[1][0:10]
```

```python
DT.head().groupby(['Pclass','Sex']).groups
```

```python
DT.head(20).groupby(['Pclass','Sex']).groups[3,'female']
```

```python
DT.head().groupby(['Pclass','Sex'],axis=1).groups
```

```python
DT.head().groupby('Sex').get_group('male')
```

```python
DT.head().groupby(['Sex','Pclass']).get_group(('female',1))
```

```python
DT.head().groupby('Age').mean()
DT.head().groupby('Age')['Fare'].agg(np.max)
DT.head().groupby('Age').agg(np.mean)
DT.dropna().astype({'Age':int}).groupby('Age').agg(np.size)
DT.dropna().groupby('Age').first()
DT.groupby('Age').count()
DT.groupby('Age').size() 
```

```python
DT.groupby('Age')['PassengerId'].count() # ignore Nan
```

```python
DT.groupby('Age')['Fare'].agg([np.sum,np.mean])
```

```python
DT.groupby('Age').agg([np.sum,np.mean])
```

```python
DT.groupby('Age').transform(lambda x : x-x.mean())
```
