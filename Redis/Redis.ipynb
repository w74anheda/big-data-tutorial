# Redis

```
 - redis-cli 
 - auth my_password
 - ping : pong
 
 String:
 - set key:value EX second NX(run if not exist)| XX(run if exist) GET
 - mset key1 value1 key2 value 2
 - get key
 - mget key1 key2 
 - incr key => if integer increment one
 - incrby key count
 - decr key
 - decrby key
 
 Hash:
 - HSET key attr1 val1 attr2 val2
 - HGET key attr2
 - HGETALL key
 - HDEL key attr2
 - HINCRBY attr2 5
 - HINCRBYFLOAT attr2 5
 - HLEN key => return count of fileds in hash
 - HVALS key => return all values
 - HKEYS key => return all keys
 
 List:
 - lpush key 01 02 03
 - rpush key 04
 - lrange key 0 -1
 - lpop key
 - rpop key
 - llen key
 
 Set:(no duplicate)(unordered)
 - SADD key val1 val2 val3
 - SRANDMEMBER key
 
 Sorted Sets:(no duplicate)(ordered)
 - ZADD key order value
 - ZRANGE users 0 -1 withscores
 
 HyperLogLog:
 - PFADD searchedkey val1 val2
 - PFCOUNT searchedkey
 
 Other:
 - persist key (no expire)
 - keys pattern(*reg* or *)
 - del key
 - getset key new value ==> return old value and update to new value
 - strlen key => return value length of key
 
 PubSub:
 - subscribe subject
 - publish subject message
 
 
 Transaction:
 - atomic (not fully acid)
 - multi
     - set user masoud
     - set pass 123
     - incr pass
   -exec
   
   
 Pipeline:
 Transactions vs Pipeline in Redis.
 We have seen that pipeline is a way to send all commands together from the client-side, the same applies to Redis transactions as well, then how are they different?

    The difference is pipelines are not atomic whereas transactions are atomic, meaning 2 transactions do not run at the same time, whereas multiple pipelines can be executed by Redis-server at the same time in an interleaved fashion.
    URL: https://buildatscale.tech/what-is-redis-pipeline/
   
   
 

```

```
 - FT.CREATE
 - FT.SEARCH
 - FT.DROPINDEX
 - FT.LIST
 
 - 
    HSET cars#a1 name 'fast car' color red year 1950 
    HSET cars#b1 name 'car' color red year 1960 
    HSET cars#c1 name 'old car' color blue year 1970 
    HSET cars#d1 name 'new car' color blue year 1990 
    
- 
    FT.CREATE idx:cars ON HASH PREFIX 1 cars#
        SCHEMA name TEXT color TAG year numeric
        
    FT.CREATE idx:cars ON HASH PREFIX 2 cars# trucks#
        SCHEMA name TEXT color TAG year numeric
    
    
- FT.SEARCH idx:cars '@name:(fast car)' 
- FT.SEARCH idx:cars '@color:{red}' 
- FT.SEARCH idx:cars '@year:[1900 1970]' 
- FT.SEARCH idx:cars '@year:[1900 1970] @color:{red}' 




--- index filed types:
    - TEXT ==> support fuzzy search, typos, ...
    - TAG  ==> used for exact string lookup
    - NUMBER
    - GEO
    - VECTOR


--- query types (number / tag / text ): 
```
