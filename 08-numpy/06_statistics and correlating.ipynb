```python
# positive correlating
x = np.random.randint(0,50,1000)
y = x + np.random.normal(0,5,1000)
correlating = np.corrcoef(x,y)
print(correlating)

plt.scatter(x, y, label='Positive Correlating', alpha=0.5, c='green')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc='upper left')
plt.show()
```

```python
# nagative correlating

x = np.random.randint(0,50,1000)
y = 100 - x + np.random.normal(0,5,1000)
correlating = np.corrcoef(x,y)
print(correlating)

plt.scatter(x, y, label='Nagative Correlating', alpha=0.5 ,c='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc='upper right')
plt.show()
```

```python
# None correlating

x = np.random.randint(0,50,1000)
y = np.random.randint(0,50,1000)
correlating = np.corrcoef(x,y)
print(correlating)

plt.scatter(x, y, label='None Correlating', alpha=0.3 ,c='blue')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc='upper right')
plt.show()

```

```python
# correlating matrix

x = np.random.randint(0,50,1000)
y = x + np.random.normal(0,5,1000)
r = 100 - x + np.random.normal(0,5,1000)
t = np.random.randint(0,50,1000)

correlating = np.corrcoef([x,y,r,t])




vegetables = ["x", "y", "r", "t"]
farmers = ["x", "y", "r", "t"]



fig, ax = plt.subplots()
im = ax.imshow(correlating,vmin=-1, vmax=1,cmap='Greys')
ax.set_xticks(np.arange(len(farmers)), labels=farmers)
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)

cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel('Cbar Label', rotation=-90, va="bottom")
```

```python
dt = pd.DataFrame({
    'x' : np.random.randint(0,50,1000),
})
dt['y'] = dt['x']+ np.random.normal(0,5,1000)
dt['r'] = 100 - dt['x']+ np.random.normal(0,5,1000)
dt['t'] = np.random.randint(0,50,1000)

print(dt.corr()) # correlating

pd.plotting.scatter_matrix(dt,figsize=(9,9),color='red',alpha=0.4)
```
