```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import time
```

```python
BASE_URL = 'http://saha.moi.ir/'
USERNAME = '1741995108'
PASSWORD = '4NpzuFqdEyCDv6T'
```

```python
driver = webdriver.Firefox()
# Login To Profile
driver.get(BASE_URL)
driver.find_element(By.ID, "login").send_keys(USERNAME)
driver.find_element(By.ID, "password").send_keys(PASSWORD)
driver.find_element(By.ID, "loginbtn").click()
time.sleep(5)

base_window = driver.current_window_handle

posts = []

def collect_post():
    global posts
    new = driver.find_elements(By.CLASS_NAME, "js_feed_view_more_entry_holder")
    for post in new:
        if post.get_attribute('id') not in posts:
            posts.append({
                "id": post.get_attribute('id'),
                "element" : post,
                "href" : post.find_element(
                    By.CLASS_NAME,'feed_permalink'
                    ).get_attribute('href'),
                "done" : False
            })
        
collect_post()



def like_post_comments(post):
    global base_window;
    global driver
    driver.execute_script("window.open();")
    driver.switch_to.window(driver.window_handles[1]) 
    driver.get(post['href'])
    
    
    tumbs = driver.find_elements(By.CLASS_NAME,"js_like_link_like")
    for tumb in tumbs:
        try:
            unlike = tumb.find_element(
            By.XPATH, "following-sibling::*[1]"
                ).get_attribute('style')
            if unlike == 'display: none;':
                tumb.click()
        except Exception:
            continue
    driver.close()
    driver.switch_to.window(base_window)


def handle_posts(func):
    global posts
    for post in posts:
        try:
            func(post)
        except Exception:
            continue 
        
handle_posts(like_post_comments)

driver.quit()


def data_load():
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        time.sleep(5)

        new_height = driver.execute_script("return document.body.scrollHeight")

        if new_height == last_height:
            break
        last_height = new_height 
    
    while True:
        driver.find_element(By.CLASS_NAME, "global_view_more").click()
        time.sleep(5)
        if not driver.find_element(By.CLASS_NAME, "global_view_more"):
            break;
    
# data_load()
```
